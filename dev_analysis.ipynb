```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
db = client['ai']
collection = db['promptOptions']
```

#### Fetch all data

```python
promptOptions = collection.find({}, {'promptOptions':1})

options = []
for data in promptOptions:
    options.extend(data['promptOptions'])
```

```python
from collections import Counter

option_counts = Counter(options)
option_counts_dict = dict(option_counts)
```

#### Plot Graphs

```python
import matplotlib.pyplot as plt

labels = list(option_counts_dict.keys())
values = list(option_counts_dict.values())

plt.bar(labels, values)
plt.xlabel('Options')
plt.ylabel('Usage Count')
plt.title('Usage of Options in Prompts')
plt.xticks(rotation=45)

plt.show()
```
